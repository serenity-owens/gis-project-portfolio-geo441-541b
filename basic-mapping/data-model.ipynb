<mark>[A geodatabase is something that holds geographic data. Inside the geodatabase you can have feature datasets. Feature datasets are like a folder which stores and organizes feature classes. Feature classes store geographic features such as points and lines. It is sort of like Russian dolls. The geodatabase is the biggest doll, and it has feature datasets in it (the medium doll) and then the medium doll has feature classes in it (the smallest doll). It is basically just multiple folders or containers used to help to better organize data and resources used for geographic information systems.]</mark>
