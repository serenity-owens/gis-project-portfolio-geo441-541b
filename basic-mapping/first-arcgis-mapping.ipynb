```
https://drive.google.com/file/d/14dhKvyR1Tn_xSBmyZa9uVcW8wPqK0XwK/view?usp=sharing
```
To adjust the default sharable link:
    - Keep the **`14dhKvyR1Tn_xSBmyZa9uVcW8wPqK0XwK`** part because this is the unique file ID for the picture
    - Delete **`/view?usp=sharing`** at the end of the link
    - Replace **`/file/d/`** with **`/uc?export=view&id=`**

  The direct link for view from this example would be:
```
https://drive.google.com/uc?export=view&id=14dhKvyR1Tn_xSBmyZa9uVcW8wPqK0XwK
```
